```python
#module imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import glob

os.chdir('/yourPath/')
```

```python
#Functions
#Clean WQP
def dfCols_toDict(df, main_column, dict_columns):
    """
    Create dictionaries for each column in the list using the main column as keys.

    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - main_column (str): The main column to use as keys in the dictionaries.
    - dict_columns (list): List of columns for which dictionaries will be created.

    Returns:
    - tuple: A tuple of dictionaries, each for a column in the list.
    """
    result_dicts = tuple(
        dict(zip(df[main_column].values, df[column].values)) for column in dict_columns
    )

    return result_dicts

def cleanWQP(df, site_df, date):
    #Rename and format date time/units/tss column
    df['tss'] = df['value']
    
    df.astype({'particle_size': 'str'}).dtypes
    df['date'] = pd.to_datetime(df['date'])
    df = df[(df.units == 'mg/l')]
    df=df.copy()
    #Create dictionaries of context variables to add to each value in the dataframe
    main_column = 'SiteID'
    columns_to_dict = ['SUB_AREA', 'UP_AREA', 'DIST_MAIN']

    subArea_dict, upArea_dict, distMain_dict = dfCols_toDict(site_df, main_column, columns_to_dict)
    dicts = dfCols_toDict(site_df, main_column, columns_to_dict)
    # Combine dictionaries into a single result variable
    dict_dicts = dict(zip(columns_to_dict, dicts))

    # Map dictionaries back onto the new DataFrame
    for column, dictionary in dict_dicts.items():
        df[column] = df[main_column].map(dictionary)

    #Create huc dict and Manually add AK because it is not correctly placed in WQP
    huc_dict = dict(zip(df['SiteID'].values, df['huc'].values))
    huc_dict.update({'KENAI_WQX-10000030' : 19020302.0 ,'KENAI_WQX-10000028': 19020302.0, 'KENAI_WQX-10000017': 19020302.0, 'KENAI_WQX-10000016': 19020302.0, 'KENAI_WQX-10000005': 19020302.0})
    df.loc[:,'huc'] = df['SiteID'].map(huc_dict)
    
    df = df[df['date'] > date]

    return df, huc_dict

#LS2

def siteProximity(df):
    #Add back in doubled SiteID's to get its matchups
    df2 = df.loc[:, df.columns != 'SiteID'].dropna(subset = ['SiteID2'])
    df2['SiteID'] = df2['SiteID2']
    df = pd.concat([df, df2], axis=0, ignore_index=True).drop(['SiteID2'], axis=1).copy()

    #Add back in tripled SiteID's to get its matchups
    df2 = df.loc[:, df.columns != 'SiteID'].dropna(subset = ['SiteID1'])
    df2['SiteID'] = df2['SiteID1']
    df = pd.concat([df, df2], axis=0, ignore_index=True).drop(['SiteID1'], axis=1).copy()
    
    return df


#Add band ratios
def bandRatios(df):
    df = df.assign(NR = df['nir']/df['red'],
                    BR = df['blue']/df['red'],
                    GR = df['green']/df['red'],
                    SR = df['swir1']/df['red'],
                    BG = df['blue']/df['green'],
                    RG = df['red']/df['green'],
                    NG = df['nir']/df['green'],
                    SG = df['swir1']/df['green'],
                    BN = df['blue']/df['nir'],
                    GN = df['green']/df['nir'],
                    RN = df['red']/df['nir'],
                    SN = df['swir1']/df['nir'],
                    BS = df['blue']/df['swir1'],
                    GS = df['green']/df['swir1'],
                    RS = df['red']/df['swir1'],
                    NS = df['nir']/df['swir1'],
                    R_GN = df['red']/ (df['green'] + df['nir']),
                    R_GB = df['red']/ (df['green'] + df['blue']),
                    R_GS = df['red']/ (df['green'] + df['swir1']),
                    R_BN = df['red']/ (df['blue'] + df['nir']),
                    R_BS = df['red']/ (df['blue'] + df['swir1']),
                    R_NS = df['red']/ (df['nir'] + df['swir1']),
                    G_BR = df['green']/ (df['blue'] + df['red']),
                    G_BN = df['green'] / (df['blue'] + df['nir']),
                    G_BS = df['green'] / (df['blue'] + df['swir1']),
                    G_RN = df['green'] / (df['red'] + df['nir']),
                    G_RB = df['green'] / (df['red'] + df['blue']),
                    G_NS = df['green'] / (df['nir'] + df['swir1']),
                    B_RG = df['blue'] / (df['red'] + df['green']),
                    B_RN = df['blue'] / (df['red'] + df['nir']),
                    B_RS = df['blue'] / (df['red'] + df['swir1']),
                    B_GN = df['blue'] / (df['green'] + df['nir']),
                    B_GS = df['blue'] / (df['green'] + df['swir1']),
                    B_NS = df['blue'] / (df['nir'] + df['swir1']),
                    N_RG = df['nir'] / (df['red'] + df['green']),
                    N_RB = df['nir'] / (df['red'] + df['blue']),
                    N_RS = df['nir'] / (df['red'] + df['swir1']),
                    N_GB = df['nir'] / (df['green'] + df['blue']),
                    N_GS = df['nir'] / (df['green'] + df['swir1']),
                    N_BS = df['nir'] / (df['blue']  + df['swir1']),

                    GR2 = (df['green'] + df['red']) / 2,
                    GN2 = (df['green'] + df['nir']) / 2,
                    RN2 = (df['red'] + df['nir']) / 2,

                    #blooms
                    BR_G = (df['blue'] - df['red']) / df['green'],
                    #NS_NR = (df['nir'] - df['swir1']) / (df['red'] - df['swir1']),
                    fai = df['nir'] - (df['red'] + (df['swir1']-df['red'])*((830-660)/(1650-660))),
                    # fai = (nir - red) + (red -swir) * (830-660)/(1648-660)
                    N_S= df['nir'] - df['swir1'],
                    N_R = df['nir']- df['red'],
                    #
                    ndvi = ((df['nir']-df['red'])/(df['nir']+df['red'])),
                    ndwi = ((df['green']- df['swir1'])/(df['green'] + df['swir1'])),
                    ndssi = ((df['blue'] - df['nir'])/ (df['blue'] + df['nir'])),
                    gn_gn= ((df['green']- df['nir'])/ (df['green'] + df['nir'])),
                    TZID = 'UTC')
    return df


# Define the function to apply to the new 'sat' column
def get_sat(satID):
    if satID.startswith('LT05'):
        return int(5.0)
    elif satID.startswith('LE07'):
        return int(7.0)
    elif satID.startswith('LC08'):
        return int(8.0)
    elif satID.startswith('LC09'):
        return int(9.0)
    elif satID.startswith('L2A'):
        return int(2.0)
    else:
        return 'Other'

#LS2 cleaning
def cleanLS2(df, site_df, date):
    df=df.dropna(subset='red').drop_duplicates(subset = ['SiteID', 'date', 'sceneID', 'red']) #groupby(['SiteID', 'date']).mean().reset_index()
    df = df[(df.red > 0) & (df.red < 4000) & (df.nir > 0) & (df.blue > 0) & (df.green > 0) & (df.swir1 > 0) & (df.swir2 > 0) & (df.nir > 0)]

    
    #Create dictionaries of context variables to add to each value in the dataframe
    main_column = 'SiteID'
    columns_to_dict = ['SUB_AREA', 'UP_AREA', 'DIST_MAIN', 'width_m']

    dicts = dfCols_toDict(site_df, main_column, columns_to_dict)
    # Combine dictionaries into a single result variable
    dict_dicts = dict(zip(columns_to_dict, dicts))

    # Map dictionaries back onto the new DataFrame
    for column, dictionary in dict_dicts.items():
        df[column] = df[main_column].map(dictionary)

    df = df.drop(['width_m'], axis=1)

    df['date'] = pd.to_datetime(df['date'])
    df = df[df['date'] > date]
    
    df['Matchup'] = 'LS2'
    df['sat'] = df['sceneID'].apply(lambda x: get_sat(x))
    df = bandRatios(df)
    
    tmp = df.select_dtypes(include=[np.number])
    df.loc[:, tmp.columns] = np.round(tmp, 3)
    
    return df

#Fusion cleaning
def cleanFusion(df, site_df, date):
    df=df.dropna(subset='red').drop_duplicates(subset = ['SiteID', 'date', 'red']) #groupby(['SiteID', 'date']).mean().reset_index()
    df = df[(df.red > 0) & (df.red < 4000) & (df.nir > 0) & (df.blue > 0) & (df.green > 0) & (df.swir1 > 0) & (df.swir2 > 0) & (df.nir > 0)]

    
    df = df[df['date'] > date]
    
    #Create dictionaries of context variables to add to each value in the dataframe
    main_column = 'SiteID'
    columns_to_dict = ['SUB_AREA', 'UP_AREA', 'DIST_MAIN', 'width_m']

    dicts = dfCols_toDict(site_df, main_column, columns_to_dict)
    # Combine dictionaries into a single result variable
    dict_dicts = dict(zip(columns_to_dict, dicts))

    # Map dictionaries back onto the new DataFrame
    for column, dictionary in dict_dicts.items():
        df[column] = df[main_column].map(dictionary)

    df = df.drop(['width_m'], axis=1)

    df['sat'] = 1.0
    df = bandRatios(df)
    
    #Add date/time columns
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = pd.to_datetime(df['date']).dt.year
    df['month'] = pd.to_datetime(df['date']).dt.month
    
    df['Matchup'] = 'Fusion'

    tmp = df.select_dtypes(include=[np.number])
    df.loc[:, tmp.columns] = np.round(tmp, 3)
    
    return df
    
    
#Merge data +/- 1 day
def oneDayMerge(df1, df2):
    
    #Sort and harmonize column date types
    df1 = df1.sort_values(by = 'date')
    df1['date'] = pd.to_datetime(df1['date'])
    df2 = df2[['tss', 'parameter', 'units', 'SiteID', 'particle_size', 'time', 'date', 'sample_method', 'analytical_method', 'lat', 'long']].sort_values(by = 'date')
    df2['date'] = pd.to_datetime(df2['date'])
    
    #Perform merge within 1 day before or after wqp TSS measurement date
    matchups = pd.merge_asof(df2, df1, on = 'date', by ='SiteID',
                                     tolerance=pd.Timedelta(1, "d"),
                                     allow_exact_matches=True,
                                    direction = 'nearest')
    
    #Pick one measurement per day
    matchups = matchups.dropna(subset= ['red']).drop_duplicates(subset=['SiteID', 'date', 'tss'])
    
    #Format dates and times
    datetime_str = matchups['date'].dt.strftime('%Y-%m-%d') + ' ' + matchups['time']
    matchups['date_unity'] = pd.to_datetime(datetime_str, format = "%Y-%m-%d %H:%M:%S").dt.strftime('%Y-%m-%d %H:%M:%S')
    matchups['year'] = pd.to_datetime(matchups['date']).dt.year
    matchups['month'] = pd.to_datetime(matchups['date']).dt.month
    matchups['date'] = pd.to_datetime(matchups['date'], format = '%Y-%m-%d')
    
    #Only concenctrations for this project
    matchups = matchups[(matchups.units == 'mg/l')]

    return matchups

#Unmatched
#LS2
def makeUnmatchedPredictions(raw_data, matchup_data, merge_columns):
    #Merge with indicator in
    df = pd.merge(raw_data,matchup_data[merge_columns], indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)

    return df
```
